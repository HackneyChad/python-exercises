Do any cars have better city mileage than highway mileage?

```python
# mileage numbers
# city mileage
# hwy mileage
# city greater than highway

# fields needed

(df_gas_mileage.hwy > df_gas_mileage.cty).any()
```

Create a column named mileage_difference this column should contain the difference between highway and city mileage for each car.

```python
# This defines what the new column does, or holds
mileage_difference = df_gas_mileage.hwy - df_gas_mileage.cty

# this creates the new column in the data set
df_gas_mileage['mileage_difference'] = mileage_difference

# this calls the column to print
df_gas_mileage
```

On average, which manufacturer has the best miles per gallon?

```python
# Combined fuel economy is a weighted average of City and Highway MPG values that is calculated
# by weighting the City value by 55% and the Highway value by 45%.
# (hwy * .45) + (cty * .55)
# per https://www.fueleconomy.gov/feg/label/learn-more-gasoline-label.shtml

# This defines what the new column does, or holds
combined_fuel_econ = (df_gas_mileage.hwy * .45) + (df_gas_mileage.cty * .55)

# this creates the new column in the data set
df_gas_mileage['combined_fuel_econ'] = combined_fuel_econ

# this calls the column to print
df_gas_mileage
```

```python
df_gas_mileage[['manufacturer', 'combined_fuel_econ']].groupby('manufacturer').agg(['mean'])
```

Do automatic or manual cars have better miles per gallon?

```python
# I could return mean mileage for each type of trans, but could not group all the man. & auto trans respectively.
```

```python
df_gas_mileage.mileage_by_trans = df_gas_mileage[['trans', 'combined_fuel_econ']].groupby('trans').agg(['mean'])
df_gas_mileage.mileage_by_trans
```
